### concat() - 데이터 연결

#### 행 방향으로 연결
기본 값 ``axis=0``
같은 column끼리 데이터 연결

```python
pd.concat([gas1, gas2])
```

index가 초기화되지 않아 전체 DataFrame의 개수와 index가 동일 X

```python
gas = pd.concat([gas1, gas2], ignore_index = True)
gas
```

#### 열 방향으로 연결
``axis = 1`` 지정필요

```python
gas1 = gas.iloc[:, :5]
gas2 = gas.iloc[:, 5:]
```

```python
gas1.head()
```

```python
gas2.head()
```

```python
pd.concat([gas1, gas2], axis=1)
```

### merge() - 병합
공통된 key값(컬럼)을 가지고 있을 시 병합 가능

```python
df1 = pd.DataFrame({
    '고객명': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']})
df1
```

```python
df2 = pd.DataFrame({
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']})
df2
```

### 병합하는 방법 4가지
``how``옵션값을 지정해 4가지 방식으로 병합 가능
- how : {``left``, ``right``, ``outer``, ``inner``}
- dafault값 : ``inner``

```python
# how='inner' (default)
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

```python
pd.merge(df1, df2, how='outer')
```

### 병합하려는 컬럼의 이름이 다른 경우

```python
df1 = pd.DataFrame({
    '이름': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']})
df1
```

```python
df2 = pd.DataFrame({
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']})
df2
```

```python
pd.merge(df1, df2, left_on='이름', right_on='고객명')
```

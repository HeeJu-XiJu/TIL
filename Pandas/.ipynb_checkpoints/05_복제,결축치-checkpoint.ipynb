### fillna() - 결측치 채우기

```python
df = sns.load_dataset('titanic')
df1 = df.copy()
```

```python
df1.tail()
```

```python
df1['age'] = df1['age'].fillna(700)
```

```python
df1.tail()
```

카테고리형 데이터 처리 방법
- 이미 추가된 카테고리는 바로 fillna() 가능
- 없는 카테고리는 ``add_categories``로 추가 후 가능

```python
df1['deck'].fillna('A')
```

```python
# cat은 category의 지정자
df1['deck'].cat.add_categories('No Data').fillna('No Data')
```

## 통계값으로 채우기

### 평균으로 채우기

```python
df1 = df.copy()
df1['age'].fillna(df1['age'].mean()).tail()
```

### 중앙값으로 채우기

```python
df1 = df.copy()
df1['age'].fillna(df1['age'].median()).tail()
```

### 최빈값으로 채우기

```python
df1['deck'].mode()
```

```python
# 최빈값으로 채울 때는 [0]번째 index를 지정하여 채워야 함
df1['deck'].fillna(df1['deck'].mode()[0]).tail()
```

### dropna() - NaN값이 있는 데이터 제거

```python
# NaN값이 1개라도 있으면 행 전체 삭제
df1 = df.copy()
df1.dropna()
```

기본 옵션은 ``how=any``
- ``any`` : 1개라도 NaN값이 존재 시 drop
- ``all`` : 모두 NaN값이 존재 시 drop

```python
df1.dropna(how='all')
```

```python
from IPython.display import Image
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

### 새로운 칼럼 추가

```python
df1 = df.copy()
df1.head()
```

```python
df1['VIP']=True
df1.head()
```

## 삭제
### 행(row) 삭제
행 삭제 시 index를 지정해 삭제

```python
df1.drop(1)
```

```python
# 범위 지정
df1.drop(np.arange(10))
```

```python
# fancy indexing
df1.drop([1, 3, 5, 7, 9])
```

### 열 (column) 삭제
``axis=1`` 옵션 필수

```python
df1.head()
```

```python
df1.drop('class', axis=1).head()
```

```python
df1.drop(['who', 'deck', 'alive'], axis=1)
```

``inplcae = True``삭제 내용을 바로 적용

```python
df1.drop(['who', 'deck', 'alive'], axis=1, inplace=True)
df1.head()
```

### 컬럼간 연산

```python
# family(가족) = sibsp 컬럼과 parch 컬럼의 합
df1 = df.copy()
df1['family'] = df1['sibsp'] + df1['parch']
df1.head()
```

```python
# 문자열의 합(이어붙히기)
df1['gender'] = df1['who'] + '-' + df1['sex']
df1.head()
```

round() - 소수점 자릿수

```python
df1['round'] = round(df1['fare'] / df1['age'], 2)
df1.head()
```

NaN값을 1개라도 포함하고 있다면 결과는 **NaN**

```python
df1.loc[df1['age'].isnull(), 'deck':].head()
```

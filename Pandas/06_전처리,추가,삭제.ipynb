### datetime 타입
[document](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DatetimeIndex.year.html) \
``datetime``타입은 dt접근자로 아래의 날짜속성에 접근 가능

- pandas.Series.dt.month: 월
- pandas.Series.dt.day: 일
- pandas.Series.dt.hour: 시
- pandas.Series.dt.minute: 분
- pandas.Series.dt.second: 초
- pandas.Series.dt.microsecond: micro 초
- pandas.Series.dt.nanosecond: nano 초
- pandas.Series.dt.week: 주
- pandas.Series.dt.weekofyear: 연중 몇 째주
- pandas.Series.dt.dayofweek: 요일 (월요일 0, 일요일 6)
- pandas.Series.dt.weekday: 요일 (dayofweek과 동일)
- pandas.Series.dt.dayofyear: 연중 몇 번째 날
- pandas.Series.dt.quarter: 분기

### pd.to_numeric() - 수치형 변환
numerical type이 아닌 컬럼을 numerical(수치형) 컬럼으로 변환

```python
df2.head()
```

```python
df2.info()
```

운동량 컬럼은 수치형 컬럼으로 보이지만 object 타입임

```python
pd.to_numeric(df2['운동량'])
```

```python
# 2344 position에서 에러
df2.loc[2344]
```

운동량에 수치형이 아닌 개행 (\N)이 들어가 에러 발생 \
NaN값이나 숫자로 변환이 불가능한 문자열이 존재할 때 변환 실패\
``errors=`` 옵션값을 바꾸어 해결 가능\
``errors : {'ignore', 'raise', 'coerce'}, default 'raise'``

``errors='raise'``로 지정시 에러를 일으키며 코드를 중단 (기본값)

``errors='coerce'``로 지정시 잘못된 문자열은 NaN값으로 치환

```python
pd.to_numeric(df2['운동량'], errors='coerce')
```

```python
pd.to_numeric(df2['운동량'], errors='coerce').loc[2344]
```

``errors='ignore'``로 지정시 잘못된 문자열은 변환이 되지 않고 무시\
따라서 전체 컬럼의 dtype은 **object**로 남아있음

```python
pd.to_numeric(df2['운동량'], errors='ignore')
```

```python
pd.to_numeric(df2['운동량'], errors='ignore').loc[2344]
```

```python
df2['운동량'] = pd.to_numeric(df2['운동량'], errors='coerce')
```

### pd.cut() - 구간 나누기(binning)
연속된 수치(continuous values)를 구간으로 나누어 카테고리화할 때 사용

```python
df2.head()
```

```python
# right=False로 지정시 우측 범위를 포함하지 않음
bins = [0, 200, 400, df2['운동량'].max()]
pd.cut(df2['운동량'], bins, right=False)
```

``labels``를 지정해줄 수 있으며, 지정한 bins의 개수보다 1개가 적어야 함

```python
labels = ['운동부족', '보통', '많음']
pd.cut(df2['운동량'], bins, labels=labels, right=False)
```

```python
df2.describe()
```

``bins`` 옵션에 나누고자 하는 구간의 개수를 설정

```python
df2['운동량_cut'] = pd.cut(df2['운동량'], bins=10)
```

```python
df2['운동량_cut'].value_counts()
```

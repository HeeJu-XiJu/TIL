```python
from IPython.display import Image
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

## apply() - 함수 적용

```python
df.head()
```

```python
df['who'].value_counts()
```

### 함수 (function) 정의

```python
def tranform_who(x):
    if x == 'man' :
        return '남자'
    elif x == 'woman':
        return '여자'
    else:
        return '아이'
```

```python
df['who'].apply(tranform_who)
```

```python
df['who'].apply(tranform_who).value_counts()
```

```python
def tranform_who(x):
    return x['fare'] / x['age']
```

```python
df.apply(tranform_who, axis=1)
```

## apply() - lambda 함수

```python
df['survived'].value_counts()
```

```python
df['survived'].apply(lambda x : '생존' if x == 1 else '사망').value_counts()
```

## groupbt() - 그룹

```python
df.head()
```

```python
df.groupby('sex').mean(numeric_only=True)
```

### 2개 이상의 컬럼으로 그룹
list로 묶어서 지정

```python
df.groupby(['sex', 'pclass']).mean(numeric_only=True)
```

### 1개의 특정 컬럼에 대한 결과 도출

```python
df.groupby(['sex', 'pclass'])['survived'].mean(numeric_only=True)
```

``pd.DataFrame()``, ``survived``컬럼을 ``[]``으로 한번더 감싸주는 방법

```python
pd.DataFrame(df.groupby(['sex', 'pclass'])['survived'].mean(numeric_only=True))
```

```python
df.groupby(['sex', 'pclass'])[['survived']].mean(numeric_only=True)
```

### reset_index() : 인덱스 초기화
그룹핑된 데이터프레임의 index를 초기화하여 새로운 데이터프레임 생성

```python
df.groupby(['sex', 'pclass'])['survived'].mean(numeric_only=True).reset_index()
```

### 다중 컬럼에 대한 결과도출

```python
df.groupby(['sex', 'pclass'])[['survived', 'age']].mean(numeric_only = True)
```

### 다중 통계 함수 적용
여러가지의 통계값을 적용할 때는 ``agg()``를 사용

```python
df.groupby(['sex', 'pclass'])[['survived', 'age']].agg(['mean', 'sum', 'max', 'min'])
```

### pivot_table()
``index``, ``columns``, ``values``를 지정해 피벗

### 1개의 그룹에 대한 단일 컬럼 결과

```python
df.pivot_table(index='who', values='survived')
```

```python
df.pivot_table(columns='who', values='survived')
```

### 다중 그룹에 대한 단일 컬럼 결과

```python
df.pivot_table(index=['who', 'pclass'], values='survived')
```

### index에 컬럼을 중첩하지 않고 행과 열로 펼친 결과

```python
df.pivot_table(index='who', columns='pclass', values='survived')
```

### 다중 통계함수 적용

```python
df.pivot_table(index='who', columns='pclass', values='survived', aggfunc=['sum', 'mean'])
```
